```python
sr_no = soup.find_all('td', class_ = "body_text_table")

array1 = []
for elem in sr_no:
    array1.append(elem.text.strip('\t\r\n                        '))
    
print(array1)
```

```python
array2 = np.array(array1)
array2 = np.delete(array1, [280,281])
len(array2)
```

```python
splitedSize = 5
a_splited = [array2[x:x+splitedSize] for x in range(0, len(array2), splitedSize)]

print(a_splited)
```
